```python
import requests
import os
from bs4 import BeautifulSoup
```

```python
r = requests.get('https://www.firstpost.com/')
print(r.content)
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
print(soup.prettify())
```

```python
s = soup.find('div', id= '__next')
s
lines = s.find_all('p')
for line in lines:
    op = line.text
    print(op)
```

```python
lines = s.find_all('p')

output_file = 'sv_txt.txt'
output_path = os.path.join('output', output_file)
new_out = open(output_path, 'w', encoding='utf-8')
for line in lines:
    new_out.write(f'({line.text})'+'\n')
new_out.close()
print('Successfully saved')
```
