```python
keyword_sentence_dict = {}

input_dir = 'output'
input_file_path = "sv_txt.txt"
path = os.path.join(input_dir, input_file_path)
try:
    with open(path, 'r', encoding='utf-8') as file:
        sentences = file.readlines()
        for sentence in sentences:
            for keyword in keywords:
                if keyword in sentence:
                    if keyword not in keyword_sentence_dict:
                        keyword_sentence_dict[keyword] = sentence.strip()
                        break
except FileNotFoundError:
    print(f"Error: The file '{input_file_path}' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")


for k, v in keyword_sentence_dict.items():
    print(f"{k}= {v}")
```

```python
from geopy.geocoders import Nominatim
```

```python
coords = []
for keyword in keyword_sentence_dict.keys():
    geolocator = Nominatim(user_agent="pycourse")
    location = geolocator.geocode(keyword)
    coords.append((location.latitude, location.longitude))
```

```python
descriptions=list(keyword_sentence_dict.values())
```

```python
import folium
from folium.plugins import MarkerCluster
from folium.plugins import Fullscreen
```

```python
my_map = folium.Map(
    location=[20, 0],  # Center the map on a default location
    zoom_start=2,min_zoom=2, zoom_control=True, max_bounds=True,
    tiles="CartoDB Dark Matter"  # Disable default tiles
)
southwest = [-90, -180]  # Bottom-left corner (latitude, longitude)
northeast = [90, 180]    # Top-right corner (latitude, longitude)

# Add the bounds
my_map.fit_bounds([southwest, northeast])
Fullscreen(position='topright', title='Full Screen', title_cancel='Exit Full Screen').add_to(my_map)
```

```python
custom_cluster = """
function(cluster) {
    return L.divIcon({
        html: '<div style="background-color: brown; color: white; border-radius: 50%; padding: 5px;">' + cluster.getChildCount() + '</div>',
        className: 'marker-cluster',
        iconSize: [30, 30]
    });
}
"""
marker_cluster = MarkerCluster(icon_create_function=custom_cluster).add_to(my_map)
for coord, description in zip(coords, descriptions):
    static_text = '<br><a href="https://www.firstpost.com" target="_blank" style="color: #0066cc; text-decoration: none;">Visit Firstpost</a>'
    styled_popup = folium.Popup(
        f'<div style="background-color: #eae3e3; border: 2px solid #333; border-radius: 10px; padding: 10px; color: #333; font-size: 14px;">{description}{static_text}</div>', 
        max_width=300
    )
    folium.Marker(
    location=coord,
    popup=styled_popup,
    icon=folium.Icon(color='darkred')
).add_to(marker_cluster)
my_map
```
